## Introduction

This weekend, the jailctf team hosted [jailctf](https://ctf.pyjail.club), which included a challenge inspired by this competition. This challenge required solving the hypothetical task000 listed on the overview tab in **64 bytes** or less.



### Solving

The first thing to note is that the grid has a fixed width of 10, which is key because it means that using **regex**, we can hardcode some distances. The pattern we want to match is roughly
```
__X_______
___5______
```
where `X` is a color that isn't gray (background). We want to replace the gray cell that we capture with a black cell. Ok, easy enough.

Since we want to match all the corresponding gray cells on the first passthrough, we should use a lookbehind to check for the presence of the `X` color, since lookbehinds do not consume tokens. A very basic regex script is as follows:

````python
import re
p=lambda g:eval(re.sub('(?<=[12346789].{34})5','0',str(g)))
````

This basic example is already 69 bytes long, and doing some basic regex golfing on it, we can get it down to **64 bytes**. If you have any optimizations to this, let me know in the comments.

````python
import re
p=lambda g:eval(re.sub('(?<=[^5],.{33})5','0',str(g)))
````

Apparently there's <= 64 byte non-regex solutions (60 bytes!) using secret tech also, but for this task regex works well enough :).
