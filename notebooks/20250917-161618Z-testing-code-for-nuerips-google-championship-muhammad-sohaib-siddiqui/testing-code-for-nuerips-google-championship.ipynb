```python
import json

# Load the task file
with open("/kaggle/input/google-code-golf-2025/task004.json", "r") as f:
    task = json.load(f)

# Check what’s inside
print(task.keys())     # should show: train, test, arc-gen
print(len(task["train"]), "training examples")
```

```python
# First training example
train_example = task["train"][0]

print("Input grid:")
for row in train_example["input"]:
    print(row)

print("\nOutput grid:")
for row in train_example["output"]:
    print(row)
```

```python
import numpy as np
import matplotlib.pyplot as plt

def show_grid(grid, title=None):
    arr = np.array(grid)
    plt.imshow(arr, cmap="tab10", vmin=0, vmax=9)  # 0–9 → 10 colors
    plt.axis("off")
    if title:
        plt.title(title)

# Show input and output side by side
inp = train_example["input"]
out = train_example["output"]

plt.figure(figsize=(6,3))
plt.subplot(1,2,1); show_grid(inp, "Input")
plt.subplot(1,2,2); show_grid(out, "Expected Output")
plt.show()
```
