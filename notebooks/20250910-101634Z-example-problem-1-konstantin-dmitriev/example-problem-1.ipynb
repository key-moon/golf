# Посмотрим, как выглядят входные и выходные данные

```python
inp, out = examples['train'][0]['input'], examples['train'][0]['output']

def print_matrix(inp):
    x = [str(n) for n in range(len(inp[0]))]
    print(f'    {"  ".join(x)}')
    for y,r in enumerate(inp):
        print(f'{y:2}', r)

show_examples(examples['train'][:1])

print('Input')
print_matrix(inp)

print('\nOutput')
print_matrix(out)
```

# Предложим решение

```python
def p(inp):
    out = []
    for y in range(9):
        row = []
        for x in range(9):
            z1 = inp[y // 3][x // 3]
            z2 = inp[y % 3][x % 3]
            z = z1 & z2
            row.append(z)
        out.append(row)
    return out
```

```python
print('p(Input)')
print_matrix(p(inp))

print('\nOutput')
print_matrix(out)

print('\nSuccess!' if p(inp) == out else 'Failure :(')
```
